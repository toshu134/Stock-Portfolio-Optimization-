```python
#drop date column to make data processing easier   
df=pd.read_csv(file_path)
df=df.drop('Date',axis=1)
```

```python
#converting the string into float 
for col in df.columns:
    if df[col].dtype=='object':
        print("aand")
        df[col]=df[col].str.replace(',', '').astype('float64')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
l = [col for col in df.columns]

trans = ColumnTransformer(
    transformers=[
        ('scaler', StandardScaler(), l)  # Apply StandardScaler to all columns in `l`
    ],
    remainder='passthrough'  # Keep any columns not listed in `l` as-is (if any)
)

df_scaled = trans.fit_transform(df)

df_scaled = pd.DataFrame(df_scaled, columns=l)

print(df_scaled.head())
```
